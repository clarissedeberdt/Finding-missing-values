<h2>1. Load the Dataset</h2>

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
print('Basic information')
print(df.info())

print('Summary statistics')
print(df.describe())
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
print('Missing values per columns:')
print(missing_values)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing values')

plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_values_emp = df['Employment'].isnull().sum()
print(f'Total number of missing values for Employment column : {missing_values_emp}')
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
most_frequent_emp = df['Employment'].mode()[0]
print(f'The most frequent value in Employment column are : {most_frequent_emp}')
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
most_frequent_value = df['Employment'].mode()[0]
print(f"The most frequent value in 'Employment' are: {most_frequent_value}")

df['Employment'] = df['Employment'].fillna(most_frequent_value)
missing_values = df['Employment'].isnull().sum()
print(f"Missing values in 'Employment' after imputation are : {missing_values}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
import seaborn as sns
import matplotlib.pyplot as plt

top_5_employment = df['Employment'].value_counts().nlargest(5)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_5_employment.values, y=top_5_employment.index)

plt.title('Top 5 des statuts d\'emploi')
plt.xlabel('Nombre')
plt.show()
```
